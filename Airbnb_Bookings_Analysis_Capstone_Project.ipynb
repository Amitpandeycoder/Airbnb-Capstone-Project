## **Sorting the Data & seperate the useful column from the Data**

```python
df1 = df[['id','name','host_id','host_name','neighbourhood_group','neighbourhood','room_type','price','minimum_nights',
             'number_of_reviews','calculated_host_listings_count','availability_365']]
```

```python
df1.head(5)
```

##1. What can we learn about different hosts and areas?

```python
hosts_areas = df1.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
hosts_areas.sort_values(by='calculated_host_listings_count', ascending=False).head(5)
```



##As you can see high number of host_listings are from **Manhattan** created Sonder (NYC), Blueground, Michael, David	

##2. What can we learn from predictions? (ex: locations, prices, reviews, etc)

```python
areas_reviews = df1.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
areas_reviews
```

Import **matplotlib. pyplot** in the notebook. This is a collection of command style functions that make matplotlib work like **MATLAB**. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc. 

```python
import matplotlib.pyplot as plt
```

```python
area = areas_reviews['neighbourhood_group']
review = areas_reviews['number_of_reviews']

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(area, review, color ='maroon',
        width = 0.4)
 
plt.xlabel("area")
plt.ylabel("review")
plt.title("Area vs Number of reviews")
plt.show()
```
